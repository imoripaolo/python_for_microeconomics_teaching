```python
"""

Ilustrating the UMP / Ilustrando o PMU

Author / Autor: imoripaolo

"""

import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
"""
The Utility Maximization Problem:

max_{x_1,x_2} u(x_1,x_2) st x_1p_1 + x_2p_2 = w

(We'll consider the case where the BC is valid with inequality later.)

Note that from the BC, we have:

x_2 = w/p_2 - (p_1/p_2)x_1

Therefore, we can write the UMP as:

max_{x_1} u(x_1, w/p_2 - (p_1/p_2)x_1)

Method: Grid search; let's also draw the Budget Constraint and some indifference curves.

-----

O Problema de Maximização de Utilidade:

max_{x_1,x_2} u(x_1,x_2) st x_1p_1 + x_2p_2 = w

(Vamos considerar o caso em que a restrição é válida com desigualdade depois.)

Note que a partir da restrição, temos que:

x_2 = w/p_2 - (p_1/p_2)x_1

Logo, podemos escrever o PMU como

max_{x_1} u(x_1, w/p_2 - (p_1/p_2)x_1)

Method: Procura no grid; vamos também desenhar a RO e algumas curvas de indiferença

"""

# From the Budget Constraint Script / Do script de Restrição Orçamentária:
w,p1,p2 = 100, 5, 5
x1_min, x1_max = 0.1, w/p1
x1 = np.linspace(x1_min,x1_max, 100)
x2 = (w/p2) - (x1*p1)/p2

# From the Indifference Map Script / Do script de Mapas de Indiferença:
def cd(x,y):
    output = (x**0.6)*(y**0.4)
    return output

def ind(x,u):
    output = (u/(x**0.6))**(1/0.4)
    return output

# We'll now do the grid search for each value of x_1:
# Vamos agora fazer procura no grid para cada valor de x_1:

u = np.zeros(100)
for i in range(0,100):
    u[i] = cd(x1[i], (w/p2) - (x1[i]*p1)/p2)

fig, ax = plt.subplots()
ax.plot(u,'b')
ax.set_ylabel('u')
ax.set_xlabel('x1')
plt.show()

star = np.argmax(u)
print("As demandas Marshallianas são x_1(p,w)=", x1[star], "e x_2(p,q)=", (w/p2) - (x1[star]*p1)/p2 )
```

```python
# Now, let's plot the indifference curve and the budget constraint:
# Agora, vamos plotar a curva de indiferença e a restrição orçamentária:

i1, i0, i2 = np.zeros(100), np.zeros(100), np.zeros(100)
for i in range(0, 100):
    i1[i] = ind(x1[i],cd(x1[star], (w/p2) - (x1[star]*p1)/p2))
    i0[i] = ind(x1[i],cd(x1[star], (w/p2) - (x1[star]*p1)/p2)-1)
    i2[i] = ind(x1[i],cd(x1[star], (w/p2) - (x1[star]*p1)/p2)+1)

fig, ax = plt.subplots()
plt.xlim([0, 20])
plt.ylim([0, 20])
ax.plot(x1,x2,'k')
ax.plot(x1,i1,'b')
ax.plot(x1,i0,'b')
ax.plot(x1,i2,'b')
ax.set_ylabel('x2')
ax.set_xlabel('x1')
plt.show()
```
