```python
# Modelo de Equilíbrio Competitivo - Formação de Estoques e Ajuste de Preços

""""
Autor: imoripaolo

Baseado em Chiang e Wrainwright, 17.5

"""""

import numpy as np
import matplotlib.pyplot as plt
```

```python
""""
Formação de Estoques

1-) Demanda depende do preço corrente.

2-) Oferta depende do preço corrente.

3-) Preço evolui conforme há formação de estoques


Oferta : q_t = -10 + 5*p_t

Demanda : x_t = 200 - 10*p

Estoques: s_t = q_t - x_t

Regra de Ajuste de Preços: p_(t+1) = p_t - 0.1*s_t

"""""

# Vetores de Preços, e Quantidades

x = np.zeros(20)
q = np.zeros(20)
p = np.zeros(20)
s = np.zeros(20)

# Vamos ver para o primeiro período

p[0] = 10
q[0] = -10 + 5*p[0]
x[0] = 200 - 10*p[0]
s[0] = q[0] - x[0]
p[1] = p[0] - 0.1*s[0]

p[0],q[0],x[0],s[0],p[1]
```

```python
for t in range(1,19):
    q[t] = -10 + 5*p[t]
    x[t] = 200 - 10*p[t]
    s[t] = q[t] - x[t]
    p[t+1] = p[t] - 0.1*s[t]

q[19] = -10 + 5*p[19]
x[19] = 200 - 10*p[19]
s[19] = q[19] - x[19]
```

```python
fig, ax = plt.subplots()
ax.plot(p,'b')

fig, ax = plt.subplots()
ax.plot(q,'r')
ax.plot(x,'b')

fig, ax = plt.subplots()
ax.plot(s,'b')
```
