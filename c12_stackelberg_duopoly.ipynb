```python
""""

Model of Stackelberg Duopoly / Modelo de Duopólio de Stackelberg

Author / Autor: imoripaolo

Based of / Baseado em: Tadelis, sec 8.3.2

"""""
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
"""
The Stackelberg Duopoly: firm 1 plays first and chooses q_1. After that, firm 2 is called to play
and chooses q_2 based on q_1. Therefore, firm 1 chooses q_1 based on the reaction of firm 2.

Market Demand: p = 100 - q

Market Supply: q = q_1 + q_2

Cost: c(q_j) = 10q_j, j = 1,2;

Problem of firm 2: max_{q_2} (100 - q_1-q_2)q_2 - 10 q_2

FOC: 100 - q_1 - 2q_2 - 10 = 0

Best Response: q_2^* = (90-q_1)/2

Problem of firm 1: max_{q_1} (100 - q_1-q_2^*)q_1 - 10 q_1

FOC: 100 - 2q_1 - q_2^* - 10 = 0

Best Response: ?
-----

Duopólio de Stackelberg: firma 1 joga antes e escolhe q_1. Após isso, a firma 2 é chamada a jogar e
escolhe q_2 baseado em q_1. Logo, firma 1 escolhe q_1 baseado na reação da firma 2.

Demanda de Mercado: p = 100 - q

Oferta de Mercado: q = q_1 + q_2

Custo: c(q_j) = 10q_j, j = 1,2;

Problema da firma 2: max_{q_2} (100 - q_1-q_2)q_2 - 10 q_2

CPO: 100 - q_1 - 2q_2 - 10 = 0

Best Response: q_2^* = (90-q_1)/2

Problema da firma 1: max_{q_1} (100 - q_1- (90-q_1)/2)q_1 - 10 q_1

CPO: 100 - 2q_1 - 45 - q_1 - 10 = 0

Best Response: ?

"""

def BR_stack(x):
    BR = (1/2)*(90-x)
    return BR

def price(x):
    p = 100 - x
    return p

def profit(x1,x2):
    pi = (100 - x1 - x2 - 10)*x1 
    return pi
```

```python
"""
Problem of the firm 2:

We check the best reponse of firm 2 for each choice of quantity by firm 1. We'll plot the best 
response q_2 and the profit associated with each BR.

---

Problema da firma 2:

Vamos verificar qual a best response da firma 2 para cada escolha de quantidade da firma 1. Vamos
plotar a best response q_2 e o lucro associado a cada BR.

"""

q, q2, pi2 = np.zeros(90), np.zeros(90), np.zeros(90)
for i in range(0,90):
    q[i] = i+1
    q2[i] = BR_stack(q[i])
    pi2[i] = profit(q2[i],q[i])

fig, ax = plt.subplots()
ax.plot(q2,'b')
ax.set_ylabel('q2')
ax.set_xlabel('q1')
plt.show()

fig, ax = plt.subplots()
ax.plot(q2,pi2,'b')
ax.set_ylabel('pi2')
ax.set_xlabel('q2')
plt.show()
```

```python
"""
Problem of the firm 1:

Given the reaction of firm 2, we select the action q_1 that maximizes profit for firm 1.

---

Problema da firma 1:

Dada a reação da firma 2, escolhemos a ação q_1 que maximiza lucros para firma 1.

"""


pi1 = np.zeros(90)
for i in range(0,90):
    pi1[i] = profit(q[i],q2[i])

fig, ax = plt.subplots()
ax.plot(q, pi1,'b')
ax.set_ylabel('pi1')
ax.set_xlabel('q1')
plt.show()

q1 = np.argmax(pi1)
# print("The quantity chosen by firm 1 is q_1=", q1)
print("A quantidade escolhida pela firma 1 é q_1=", q1)
```

```python
"""
Market Equilibrium:

Let's check some aspects of the market equilibrium of the model.

---

Equilíbrio de Mercado:

Vamos ver alguns aspectos de equilíbrio de mercado do modelo.

"""

print("As ofertas individuais são q_1=", q1, "e q_2=", q2[q1])
print('\n')
print("Os lucros são pi_1=", pi1[q1], "e pi_2=", pi2[q1])
```
