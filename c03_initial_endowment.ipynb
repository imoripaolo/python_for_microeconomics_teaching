```python
"""

Ilustrating Endowments / Ilustrando Dotações

Author / Autor: imoripaolo

"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""

Initial Endowments: consumers may want to buy or sell goods based on the prices. We'll assume a 
Cobb-Douglas utility function: 

u(x_1,x_2) = x_1^{alpha_1}x_2^{alpha_2}

The Marshallian Demands are given by:

x_1(p,w) = (alpha_1/alpha_1+alpha_2)*w/p_1

x_2(p,w) = (alpha_2/alpha_1+alpha_2)*w/p_2

His income is given by an endowment:

w = omega_1p_1 + omega_2p_2

His net demand is therefore given by:

x_1^* = x_1(p,w) - omega_1

x_2^* = x_2(p,w) - omega_2

Consumer can be a net buyer or net seller for certain price levels.

----

Dotações Iniciais: consumidores podem querer comprar ou vender bens baseados nos preços. Nós supomos
uma função de utilidade Cobb-Douglas:

u(x_1,x_2) = x_1^{alpha_1}x_2^{alpha_2}

As Demandas Marshallinas são dadas por:

x_1(p,w) = (alpha_1/alpha_1+alpha_2)*w/p_1

x_2(p,w) = (alpha_2/alpha_1+alpha_2)*w/p_2

Sua renda é dada por uma dotação:

w = omega_1p_1 + omega_2p_2

Sua Demanda Líquida é dada por:

x_1^* = x_1(p,w) - omega_1

x_2^* = x_2(p,w) - omega_2

Consumidor pode ser comprador ou vendedor líquido do bem a depender dos preços.

"""

p1, p2 = np.linspace(0.1,10,100), np.linspace(0.1,10,100)
p1_a, p2_a = 5, 5
omega_1, omega_2 = 10, 7.5
alpha_1, alpha_2 = 0.4, 0.6
x1, x2, x1net, x2net = np.zeros(100), np.zeros(100), np.zeros(100), np.zeros(100)
for i in range(0,100):
    x1[i] = (alpha_1/ (alpha_1+alpha_2))*(p1[i]*omega_1 + p2_a*omega_2)/p1[i]
    x2[i] = (alpha_2/ (alpha_1+alpha_2))*(p1_a*omega_1 + p2[i]*omega_2)/p2[i]
    x1net[i] = x1[i] - omega_1
    x2net[i] = x2[i] - omega_2

fig, ax = plt.subplots()
plt.xlim([0, 10])
plt.ylim([-5, 20])
ax.plot(p1,x1net,'b')
ax.plot(p1,x1,'r')
ax.set_ylabel('x1')
ax.set_xlabel('p1')
plt.show()

fig, ax = plt.subplots()
plt.xlim([0, 10])
plt.ylim([-5, 20])
ax.plot(p2,x2net,'b')
ax.plot(p2,x2,'r')
ax.set_ylabel('x2')
ax.set_xlabel('p2')
plt.show()
```
