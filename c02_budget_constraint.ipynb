```python
"""

Ilustrating the Budget Constraint / Ilustrando a Restrição Orçamentária

Author / Autor: imoripaolo

"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
""""
Our objective is to take a simple of a walrasian budget constraint, explore some aspects of it
and see some easy comparative statics. This ilustrates the typical undergrad problem, with 2 goods

The budget constraint is given by:

x1p1 + x2p2 = w

Let's isolate x2 in this expression:

x2 = w/p2 -x1p1/p2

----
Nosso objetivo é ver um caso simples de restrição orçamentária walrasiana, explorar alguns aspectos e
ver algumas estáticas comparativas mais simples. Este ilustra o problema típico de graduação, com
2 bens.

A restrição orçamentária é dada por:

x1p1 + x2p2 = w

Vamos isolar x2 na expressão:

x2 = w/p2 -x1p1/p2

"""""


# Let's define wealth and prices / Vamos definir renda e preços
w,p1,p2 = 100, 5, 5

# Now, we define min and max values that each good can assume
# Agora, vamos definir valores min e max que cada bem pode assumir
x1_min, x1_max = 0, w/p1
x2_min, x2_max = 0, w/p2
# Why these values? / Por que esses valores?

# Now, we create a vector of values for x1 and x2 between their min and max values:
# Agora, criamos um vetor de valores para x1 e x2 entre seus valores min e max:
x1 = np.linspace(x1_min,x1_max, 100)
x2 = (w/p2) - (x1*p1)/p2

# Now, let's plot the budget contraint / Agora vamos plotar a restrição orçamentária
fig, ax = plt.subplots()
ax.plot(x1,x2,'b')
ax.set_ylabel('x2')
ax.set_xlabel('x1')
plt.show()
```

```python
""""
Comparative statics:

We can check what happens when we change the wealth or the prices. From an initial state 
(w, p1, p2) = (100, 5, 5), let's check what happens.

----
Estáticas Comparativas:

Vamos ver o que ocorre quando alteramos a renda ou preços. A partir de um estado inicial
(w, p1, p2) = (100, 5, 5), vamos ver o que ocorre.

"""""

# CS 1: Rise in Wealth / EC 1: Aumento na Renda
w1 = 150

# The new budget constraint / A nova restrição orçamentária:
x11_min, x11_max = 0, w1/p1
x11 = np.linspace(x11_min,x11_max, 100)
x21 = (w1/p2) - (x11*p1)/p2 

# Let's plot the old and the new BC / Vamos plotar a antiga e a nova RO:
fig, ax = plt.subplots()
ax.plot(x1,x2,'b')
ax.plot(x11,x21,'r',label='EC: aumento em w')
ax.set_ylabel('x2')
ax.set_xlabel('x1')
plt.legend(loc='center right')
plt.legend()
plt.show()
```

```python
# CS 2: Equal Rise in Both Prices / EC 2: Aumento Igual em Ambos os Preços:
p12, p22 = 7.5, 7.5

# The new budget constraint / A nova restrição orçamentária:
x12_min, x12_max = 0, w/p12 
x12 = np.linspace(x12_min,x12_max, 100)
x22 = (w/p22) - (x12*p12)/p22 

# Let's plot the old and the new BC / Vamos plotar a antiga e a nova RO:
fig, ax = plt.subplots()
ax.plot(x1,x2,'b')
ax.plot(x12,x22,'r',label='EC: aumento igual em p1 e p2')
ax.set_ylabel('x2')
ax.set_xlabel('x1')
plt.legend(loc='center right')
plt.legend()
plt.show()
```

```python
# CS 3: Rise in Only One Price / EC 3: Aumento em Apenas um Preço:
p13, p23 = 10, 5

# The new budget constraint / A nova restrição orçamentária:
x13_min, x13_max = 0, w/p13 
x13 = np.linspace(x13_min,x13_max, 100)
x23 = (w/p23) - (x13*p13)/p23 

# Let's plot the old and the new BC / Vamos plotar a antiga e a nova RO:
fig, ax = plt.subplots()
ax.plot(x1,x2,'b')
ax.plot(x13,x23,'r',label='EC: aumento em p1')
ax.set_ylabel('x2')
ax.set_xlabel('x1')
plt.legend(loc='center right')
plt.legend()
plt.show()
```

```python
# CS 4: Different Change in Both Prices / EC 4: Mudanças Diferentes em Ambos os Preços
p14, p24 = 7.5, 3
# For instance, due to taxes / Por exemplo, devido a impostos 

# The new budget constraint / A nova restrição orçamentária:
x14_min, x14_max = 0, w/p14 
x14 = np.linspace(x14_min,x14_max, 100)
x24 = (w/p24) - (x14*p14)/p24 

# Let's plot the old and the new BC / Vamos plotar a antiga e a nova RO:
fig, ax = plt.subplots()
ax.plot(x1,x2,'b')
ax.plot(x14,x24,'r',label='EC: tarifa em p1, subsídio em p2')
ax.set_ylabel('x2')
ax.set_xlabel('x1')
plt.legend(loc='center right')
plt.legend()
plt.show()
```
