```python
""""

Ilustrating Global Returns to Scale / Ilustrando Retornos Globais de Escala

Author / Autor: imoripaolo

"""""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Our objective is to illustrate the difference of functions with decreasing, increasing and
constant returns to scale.

----
Nosso objetivo é ilustrar a diferença de funções com retornos de escala decrescentes, crescentes e
constantes.

"""
alpha, beta, gamma = 0.5, 0.4, 0.6

def dec(x,y):
    output = (x**alpha)*(y*beta)
    return output

def cte(x,y):
    output = (x**alpha)*(y*(1-alpha))
    return output

def inc(x,y):
    output = (x**alpha)*(y*gamma)
    return output
```

```python
# Let's make a simple vector of inputs / Vamos fazer um vetor simples de insumos:

X = np.linspace(1,100, 100)
Y = np.linspace(10,1000, 100)

# Now, let's create a vector of output in each case
# Agora, vamos criar um vetor de produto em cada caso
DEC, CTE, INC = np.zeros(100), np.zeros(100), np.zeros(100)
for i in range(0,100):
    DEC[i], CTE[i], INC[i] = dec(X[i],Y[i]), cte(X[i],Y[i]), inc(X[i],Y[i])

fig, ax = plt.subplots()
ax.plot(DEC,'b')
ax.plot(CTE,'r')
ax.plot(INC,'g')
plt.show()
```
