```python
# Modelo de Equilíbrio Competitivo - Teia de Aranha

""""
Autor: imoripaolo

Baseado em Chiang e Wrainwright, 17.4

"""""

import numpy as np
import matplotlib.pyplot as plt
```

```python
""""
A Teia de Aranha:

1-) Demanda depende do preço corrente.

2-) Oferta depende do preço defasado.

3-) Preço evolui de forma a fazer com que a oferta seja igual a demanda.

- Vamos testar duas inclinações para a curva de oferta.

Oferta 1 : q1 = -10 + 5*p

Oferta 2 : q2 = -10 + 15*p

Demanda : x = 200 - 10*p

"""""

x1 = np.zeros(40)
x2 = np.zeros(40)
q1 = np.zeros(40)
q2 = np.zeros(40)
p1 = np.zeros(41)
p2 = np.zeros(41)
```

```python
""""
Processo Iterativo:

1-) Definimos p_t

2-) p_t define q_{t+1}

3-) Equilíbrio de mercado: q_{t+1} =x_{t+1}

4-) Preço se ajusta de modo ao mercado se equilibrar.

"""""

#Vamos começar com a primera função de oferta

p1[0] = 10
q1[0] = -10 + 5*p1[0]
x1[0] = q1[0]
p1[1] = (200-q1[0])/10

p1[0],p1[1]
```

```python
for t in range (1,40):
    q1[t] = -10 + 5*p1[t-1]
    x1[t] = q1[t]
    p1[t+1] = (200-q1[t])/10
    
fig, ax = plt.subplots()
ax.plot(p1,'b')

fig, ax = plt.subplots()
ax.plot(q1,'r')
```

```python
#Vamos fazer agora a segunda função de oferta:

p2[0] = 10

for t in range (1,40):
    q2[t] = -10 + 15*p2[t-1]
    x2[t] = q2[t]
    p2[t+1] = (200-q2[t])/10
    
fig, ax = plt.subplots()
ax.plot(p2,'b')

fig, ax = plt.subplots()
ax.plot(q2,'r')
```
