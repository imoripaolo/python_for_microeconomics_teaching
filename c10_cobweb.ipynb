```python
""""

Model of Competitive Equilibrium - Cobweb / Modelo de Equilíbrio Competitivo - Teia de Aranha

Author / Autor: imoripaolo

Based of / Baseado em: Chiang e Wrainwright, 17.4

"""""

import numpy as np
import matplotlib.pyplot as plt
```

```python
""""
The Web / A Teia de Aranha:

1-) Demand is a function of current prices / Demanda depende do preço corrente.

2-) Supply is a function of lagged prices /  Oferta depende do preço defasado.

3-) Price changes in order to clear markets / Preço evolui de forma a fazer com que a 
    oferta seja igual a demanda.

- We'll test two parametrizations for the supply curve / Vamos testar duas inclinações para a 
curva de oferta.

Supply / Oferta 1 : q1 = -10 + 5*p

Supply / Oferta 2 : q2 = -10 + 15*p

Demand / Demanda : x = 200 - 10*p

"""""

x1 = np.zeros(40) # We'll creat zero vectors for supply, demand and price in each (supply) case
x2 = np.zeros(40) # Vamos criar vetores nulos para oferta, demanda e preços em cada caso (de oferta)
q1 = np.zeros(40)
q2 = np.zeros(40)
p1 = np.zeros(41)
p2 = np.zeros(41)
```

```python
""""
The timing of the model / O Timing do Modelo

1-) We define p_t / Definimos p_t

2-) p_t determines q_(t+1) / p_t define q_(t+1)

3-) Market Clears / Equilíbrio de mercado: q_(t+1) = x_(t+1)

4-) p_(t+1) is determined based on x_(t+1)

"""""

# We'll start with the first parametrization of the supply curve
# Vamos começar com a primera função de oferta

# Basic example: let's do for t=1 (entrace 0)
# Exemplo básico: vamos fazer para t=1 (entrada 0)

p1[0] = 10
q1[0] = -10 + 5*p1[0]
x1[0] = q1[0] # Note: is this step necessary? / Nota: esse passo é necessário?
p1[1] = (200-x1[0])/10

p1[0],p1[1]
```

```python
# for Loop: let's fill all the vectors
# Loop for: vamos preencher todas as entradas

for t in range (1,40):
    q1[t] = -10 + 5*p1[t]
    x1[t] = q1[t]
    p1[t+1] = (200-x1[t])/10

# Now, for some graphs
# Agora, alguns gráficos
    
fig, ax = plt.subplots()
ax.plot(p1,'b')

fig, ax = plt.subplots()
ax.plot(q1,'r')
```

```python
# Let's do now the second parametrization. What do we expect the result to be?
# Vamos fazer agora a segunda função de oferta. Qual o resultado que esperamos que aconteça?

p2[0] = 10

for t in range (0,40):
    q2[t] = -10 + 15*p2[t]
    x2[t] = q2[t]
    p2[t+1] = (200-x2[t])/10
    
fig, ax = plt.subplots()
ax.plot(p2,'b')

fig, ax = plt.subplots()
ax.plot(q2,'r')
```
