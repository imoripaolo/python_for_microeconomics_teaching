```python
"""

Long Run Equilibrium / Equilíbrio de Longo Prazo

Author / Autor: imoripaolo

Based of / Baseado em: MWG, 10.F.2

"""

import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
"""

Long Run Equilibrium / Equilíbrio de Longo Prazo

Market Demand / Demanda de Mercado: x(p) = A - Bp

Cost Function / Função de Custo: c(q) = K + alpha*q + beta*q^2

(a) Calculate the long-run competitive equilibrium price, output per firm, aggregate output, and
number of firms. Ignore the integer constraint on the number of firms. How does each of these vary
with A?

-----

Market Clearing / Equilíbrio de Mercado: x(p) = J*q(p)

Marginal Cost / Custo Marginal: MC(q) = alpha + 2beta*q

Average Cost / Custo Médio: AC(q) = (1/q)(K + alpha*q + beta*q^2)

"""

A0, B, K, alpha, beta = 1000,  1, 10, 1, 0.2 

q = np.linspace(1,15, 100)
C, MC, AC  = np.zeros(100), np.zeros(100), np.zeros(100)

for i in range(0,100):
    C[i] = K + alpha*q[i] + beta*q[i]**2
    MC[i] = alpha + 2*beta*q[i]
    AC[i] = C[i]/q[i]
```

```python
# Finding bar(q) / Achando bar(q)

fig, ax = plt.subplots()
ax.plot(q, MC,'b',label='MC')
ax.plot(q, AC,'r', label='AC')
plt.legend()
plt.show()

barq = q[np.argmin(AC)]
print("barq = ", barq )
```

```python
# Finding p = AC(bar(q)) and x(p) / Achando p = AC(bar(q)) e x(p)

p = (K + alpha*barq + beta*barq**2)/barq
x = A0 - B*p

print("p* = ", p, ", x(p) = ", x )
```

```python
# We found q* and p*, now we must find J* / Achamos q* e p*, resta achar J*

J = x / barq
print("J* = ", J)
```

```python
"""

We can see the value of A affects only the number of firms and market supply. Therefore, we'll test 
many values of A, based on p, and see the demand and number of firms in equilibrium.

---

Podemos ver que valores de A afetam apenas o número de firmas e oferta de mercado. Logo, vamos testar 
vários valores de A, baseado em p, e ver a demanda e o número de firmas em equilíbrio.

"""

A = np.linspace(500, 2000, 1500)
xA, jA = np.zeros(1500), np.zeros(1500)

for i in range(0,1500):
    xA[i] = A[i] - (B*p)
    jA[i] = xA[i] / barq
    
fig, ax = plt.subplots()
ax.plot(A, xA, 'b')
plt.show()

fig, ax = plt.subplots()
ax.plot(A, jA,'b')
plt.show()
```
