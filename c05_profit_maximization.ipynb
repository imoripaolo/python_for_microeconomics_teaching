```python
"""

The Problem of the Neoclassical Firm / O Problema da Firma Neoclássica

Author / Autor: imoripaolo

"""

import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
"""

Neoclassical Firm: Chooses it's production in order to maximize profits, subject to market prices
and own technology (cost function).

Market Prices : p = 10

Revenue Function: R(q) = 10*q

Cost Function: c(q) = q^2

Profit Function: pi(q) = 10q - q^2

First Order Condition: 10 - 2q = 0

Firm Production: q = 5

What we'll do is plot the R(q), c(q) and pi(q) functions

--------------

Firma Neoclássica: Escolhe o seu nível de produção de forma a maximizar lucros, sujeito aos preços 
de mercado e própria tecnologia (função de custo).

Preços de Mercado: p = 10

Função de Receita: R(q) = 10*q

Função de Custo: c(q) = q^2

Função de Lucro: pi(q) = 10q - q^2

Condição de Primeira Ordem: 10 - 2q = 0

Produção da Firma: q = 5

O que nós faremos é plotar as funções R(q), c(q) e pi(q)
"""
q = np.linspace(0,10,100)
R, c, pi = np.zeros(100), np.zeros(100), np.zeros(100)
for i in range(0,100):
    R[i] = 10*q[i]
    c[i]= q[i]**2
    pi[i] = R[i] - c[i]
    
fig, ax = plt.subplots()
ax.plot(q, R,'b')
ax.plot(q, c,'r')
ax.plot(q, pi,'g')
ax.set_ylabel('R, c, pi')
ax.set_xlabel('q')
plt.show()
```
